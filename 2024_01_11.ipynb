```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

iris =pd.read_csv('iris.csv')
df=iris.corr()
sns.heatmap(data=df,annot=True,fmt=
'.2f',cbar=False)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_excel('data1.xlsx')

plt.plot(df['name'], df['kor'], '--', c = 'green', marker='o',label='korean')
plt.plot(df['name'], df['math'], c = 'red', marker='o', label='math')

plt.title('Grade graph')
plt.xlabel('Name')
plt.ylabel('Grade')
plt.legend() #범례 표시하려면 label 지정필요
plt.show()
```
